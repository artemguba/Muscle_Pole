```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from muscle_pole_env import MusclePoleEnv

# Создаем экземпляр среды
env = MusclePoleEnv(render_mode="human")

# Параметры симуляции
num_episodes = 100
max_steps_per_episode = 500

# Массивы для хранения данных
episode_rewards = []
episode_lengths = []

for episode in range(num_episodes):
    state, _ = env.reset()
    total_reward = 0
    steps = 0

    while True:
        # Простой контроллер: применяем случайное действие
        action = env.action_space.sample()

        # Выполняем шаг в среде
        state, reward, terminated, truncated, _ = env.step(action)

        # Обновляем награду и количество шагов
        total_reward += reward
        steps += 1

        # Проверяем условия завершения эпизода
        if terminated or truncated or steps >= max_steps_per_episode:
            break

    # Сохраняем результаты эпизода
    episode_rewards.append(total_reward)
    episode_lengths.append(steps)
    
    print(f"Episode {episode + 1}: Total reward = {total_reward}, Steps = {steps}")

# Закрываем среду
env.close()

# Визуализация результатов
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(episode_rewards, marker='o')
plt.title('Episode Rewards')
plt.xlabel('Episode')
plt.ylabel('Total Reward')

plt.subplot(1, 2, 2)
plt.plot(episode_lengths, marker='o')
plt.title('Episode Lengths')
plt.xlabel('Episode')
plt.ylabel('Length (steps)')

plt.tight_layout()
plt.show()
```
